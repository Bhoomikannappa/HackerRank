```python
#toy cost
def maximumToys(prices, k):
  prices.sort()
  count = 1
  for i in prices:
    if k>=i:
      k -= i
      count += 1
    else:
      break
  return count

n = int(input())
t = int(input())
prices = list(map(int, input().split()))
print(maximumToys(prices,t))
```
